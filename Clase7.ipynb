```python
## Loteria de Numeros
# 1. El objetivo general es captar el input del usuario usando la funcion input y chequear si concuerda con alguno de los elementos de una lista (numeros).
# 2. Usar un while para pedirle el input al usuario (que ingrese el numero a adivinar)
# 3. Chequear si en el input, el usuario ha ingresado el valor "terminar". En caso que no sea terminar entonces seguir. "terminar" es la clave para salir del programa.
# 4. Chequear si el numero ingresado por el usuario, esta en la lista de numeros validos. Si el numero esta en la lista, imprimir "Has ganado!", de lo contrario "Has perdido. Intenta de nuevo" y continuar el ciclo.

# Extra: crea una lista aleatoria para los NUMEROS_VALIDOS, es decir, cada vez que corra el programa van a ser 3 numeros aleatorios distintos.
##TIPS: La lista de numeros validos son numeros, no son strings.

NUMEROS_VALIDOS=[23,61,180];

while(True):
    lectura = input("Diga un numero o escribar Terminar para finalizar")
    if lectura == 'Terminar':
        break
    elif int(lectura) in NUMEROS_VALIDOS:
        print('Has Ganado!')
        break
    else:
        print('Has perdido. Intenta de nuevo')
        
print('programa terminado')
```

## Validación con try except


```python
import random
NUMEROS_VALIDOS=[23,61,180];
#NUMEROS_VALIDOS=[random.randint(1,200),random.randint(1,200),random.randint(1,200)]
while(True):
    lectura = input("Diga un numero o escribar Terminar para finalizar\n")
    if lectura.lower() == 'terminar':
        break
    else:
        try:
            if int(lectura) in NUMEROS_VALIDOS:
                print('Has Ganado!')
                break
            else:
                print('Has perdido. Intenta de nuevo')
        except:# bloque para manejar errores
            print('El dato ingresado no es valido, Intenta solo con números')
            
        
print('Programa terminado')
```

```python
# Con Parametros
# parametros son variables que se usan en la funcion
# argumentos son los valores que se pasan cuando se llama a la función

def mi_func3(texto,num):
    msj = '-'*num + f'\n{texto}\n'+'-'*num
    print(msj)
    
texto = input('Escribe un texto')
mi_func3(texto,25)
    
```

```python
# funcion con parametros Default (predeterminado)
def mi_func4(nombre='Sin Nombre'):
    print(f'Hola {nombre}')

mi_func4('Eduardo')
mi_func4()
```

```python
#Funcion con parametros default
def valor_futuro(vp=1, tasa=0.01, nper=1):
    vf = vp * ((1+tasa) ** nper)
    #print('el valor futuro es ',vf)
    return vf

vf=valor_futuro(100,0.08,3)
print(vf)
vf=valor_futuro(200,0.03,5)
print(vf)
vf = valor_futuro() # llamada a la funcion sin argumentos
print(vf)
vf = valor_futuro(0.08,100,3) # no se respeta la posicion de los parametros/argumentos
print(vf)
```

```python
#argumentos posicionales
vf = valor_futuro(tasa=0.08,vp=100, nper=3)
print(vf)

#vf = valor_futuro(tasa2=0.08,vpresente=100, n=3)
#print(vf)
```
