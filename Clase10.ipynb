```python
#contar la cantidad de gaps positivo.
import pandas as pd
data = pd.read_csv('C:/Users/DELL/Documents/appl.csv', usecols=['Date','Open','Close','Adj Close'])
contador = 0
data['gap_positivo']=data.Open > data['Close'].shift()
for dato in data['gap_positivo']:
     if dato == True:
        contador+=1
print(f"Total de gaps positivo {contador}")
```

```python
contador_positivos  = data.gap_positivo[data.gap_positivo==True].count()
contador_positivos
```
